```python
import pandas as pd

# Expanding the datsset for further analysis
data = {
    "Movie": ["Inception", "The Dark Knight", "Interstellar", "Tenet", "Dunkirk",
              "Memento", "The Prestige", "Batman Begins", "Insomnia", "Following"],
    "Genre": ["Sci-Fi", "Action", "Sci-Fi", "Action", "War",
              "Thriller", "Drama", "Action", "Thriller", "Noir"],
    "Rating": [8.8, 9.0, 8.6, 7.5, 7.9,
               8.4, 8.5, 8.3, 7.2, 7.5],
    "Budget": [160, 185, 165, 200, 100,
               9, 40, 150, 46, 6],
    "Revenue": [829, 1004, 677, 363, 526,
                40, 109, 374, 113, 0.2],
    "Director": ["Christopher Nolan"] * 10,
    "Year": [2010, 2008, 2014, 2020, 2017,
             2000, 2006, 2005, 2002, 1998],
    "Awards_Won": [4, 2, 1, 0, 3, 
                   1, 2, 0, 0, 0],
    "Duration_Minutes": [148, 152, 169, 150, 106,
                         113, 130, 140, 118, 70]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Displaying and checking the datasets
print(df)
```

```python
#checking for unique genre in the films
print(df['Genre'].unique())
```

```python
#checking values for each genre
print(df['Genre'].value_counts())
```

```python
#checking for high rated films having rating rate value greater or equal to 8
high_rated  = df[df['Rating'] > 8]
print(high_rated)
```

```python
#condition for the film who has genre as Action and revenue greater than 500
action_hits = df[(df["Genre"] == "Action") & (df["Revenue"] > 500)]
print(action_hits)
```

```python
# calculating profit 
df["Profit"] = df["Revenue"] - df["Budget"]
print(df[["Movie", "Profit"]])

# Sorting movies by profit
sorted_df = df.sort_values(by="Profit", ascending=False)
print(sorted_df)
```

```python
#getting the data by genre and profit values
avg_profit_by_genre = df.groupby("Genre")["Profit"].mean()
print(avg_profit_by_genre)
```

```python
#getting the data by year and rating values
avg_rating_by_year = df.groupby("Year")["Rating"].mean()
avg_rating_by_year.plot(kind="line", title="Average Ratings Over Years")
plt.show()
```

```python
# Pie chart for genre distribution
df['Genre'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Genre Distribution')
plt.ylabel('')  # Removes the default y-label for better appearance
plt.show()
```

```python
# Histogram for rating distribution
df['Rating'].plot(kind='hist', bins=5, title='Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

# Box plot for ratings by genre
sns.boxplot(x='Genre', y='Rating', data=df)
plt.title('Ratings by Genre')
plt.xticks(rotation=45)  
plt.show()
```
